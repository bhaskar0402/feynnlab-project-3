```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, adjusted_rand_score

# Simulated data with 11 features
np.random.seed(1234)
data = skin_x  
```

```python
# Function to perform k-means with multiple random restarts
def kmeans_clustering(data, n_clusters, n_init):
    kmeans = KMeans(n_clusters=n_clusters, n_init=n_init, random_state=1234)
    kmeans.fit(data)
    return kmeans.labels_, kmeans
```

```python
# Function to perform k-means clustering for stability assessment
def perform_kmeans(data, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters)
    return kmeans.fit_predict(data)
```

```python
# Store results for comparison, KMeans models, and stability scores
results = {}
kmeans_models = []
inertias = []
stability_scores = {k: [] for k in range(2, 34)}
num_replications = 10  # Number of replications for stability assessment

for n_clusters in range(2, 34):
    labels, kmeans_model = kmeans_clustering(data, n_clusters, 10)
    kmeans_models.append(kmeans_model)
    inertias.append(kmeans_model.inertia_)

    # Stability assessment
    labels_list = [perform_kmeans(data, n_clusters) for _ in range(num_replications)]
    for i in range(num_replications):
        for j in range(i + 1, num_replications):
            score = adjusted_rand_score(labels_list[i], labels_list[j])
            stability_scores[n_clusters].append(score)

    # Analyzing the clusters
    unique, counts = np.unique(labels, return_counts=True)
    cluster_distribution = dict(zip(unique, counts))


    results[n_clusters] = {
        "Cluster Distribution": cluster_distribution,
    }
```

```python
# Displaying the results
for n_clusters, result in results.items():
    print(f"Results for {n_clusters} Segments:")
    print(f"  Cluster Distribution: {result['Cluster Distribution']}")

# Creating a scree plot
plt.figure(figsize=(10, 6))
bars = plt.bar(range(2, 34), inertias, color='blue')
plt.xlabel('number of segments')
plt.ylabel('sum of within cluster distances')
plt.title('Scree Plot for K-means Clustering')
plt.xticks(range(2, 34))

for bar, inertia in zip(bars, inertias):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.15, bar.get_height() + 1, str(inertia), fontsize=10, color='black')

plt.show()
```

```python
# Calculating average stability score for each number of clusters
average_stability = {k: np.mean(v) for k, v in stability_scores.items()}

# Identifying the number of clusters with the highest stability score
optimal_clusters = max(average_stability, key=average_stability.get)

# Output results for stability scores
print("Average Stability Scores for Different Number of Clusters:")
for k, v in average_stability.items():
    print(f"  Clusters: {k}, Average Stability Score: {v:.3f}")

print(f"\nOptimal number of clusters based on stability: {optimal_clusters}")
```

```python
# Define a function to plot top N features for each class
def plot_top_n_features_for_each_class(grouped_data, top_n=10, exclude_columns=['age']):
    # Determine the number of classes
    classes = grouped_data.index.unique()
    num_classes = len(classes)
    
    # Set up the figure size dynamically based on the number of classes
    plt.figure(figsize=(5 * num_classes, 8))
    
    for i, cls in enumerate(classes, start=1):
        # For each class, sort features by their mean value in descending order, excluding specified columns
        class_means_sorted = grouped_data.loc[cls].drop(exclude_columns).sort_values(ascending=False)
        top_n_features = class_means_sorted.head(top_n)
        
        # Plotting
        plt.subplot(1, num_classes, i)
        top_n_features.plot(kind='bar')
        plt.title(f'Class {cls} - Top {top_n} Features')
        plt.xlabel('Feature')
        plt.ylabel('Average Value')
        plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()

# Plotting the top 10 features for each class in decreasing order
plot_top_n_features_for_each_class(grouped_data, top_n=10)
```

```python
# Define a function to plot top N features for each class with enhancements for color and layout
def plot_top_n_features_for_each_class_colored(grouped_data, top_n=10, exclude_columns=['age']):
    # Determine the number of classes
    classes = grouped_data.index.unique()
    num_classes = len(classes)
    
    # Set up the figure size and layout
    plt.figure(figsize=(18, 22))
    
    for i, cls in enumerate(classes, start=1):
        # For each class, sort features by their mean value in descending order, excluding specified columns
        class_means_sorted = grouped_data.loc[cls].drop(exclude_columns).sort_values(ascending=False)
        top_n_features = class_means_sorted.head(top_n)
        
        # Plotting with color
        plt.subplot(3, 2, i)
        top_n_features.plot(kind='bar', color=plt.cm.tab10(np.arange(len(top_n_features))))
        plt.title(f'Class {cls} - Top {top_n} Features')
        plt.xlabel('Feature')
        plt.ylabel('Average Value')
        plt.xticks(rotation=90)
        plt.ylim(0, top_n_features.max() * 1.1)  # Adjust y-axis to be larger based on the max value

    plt.tight_layout()
    plt.show()

# Plotting the colorful top 10 features for each class in decreasing order with specified layout
plot_top_n_features_for_each_class_colored(grouped_data, top_n=10)
```
